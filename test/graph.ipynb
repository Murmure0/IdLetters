```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données
data = pd.read_csv("data-Score.csv")


```

```python

data.info()
```

```python
# Définir la taille du graphique
ax = plt.figure(figsize=(10, 6))

# Utiliser Seaborn pour créer le barplot
# sns.barplot(data=data, x="Language", y="Precision", color="blue", label="Precision")
palette = {"F1-score-LangId" : "#90e1c1", "F1-score-XML-Roberta" : "#f1b287"}
sns.barplot(data=data, x="Language", y="F1-Score", hue="Modele", alpha=0.9, palette=palette)
# sns.barplot(data=data, x="Language", y="F1-score-LangId", color="orange", label="LangId", alpha=0.5)
# Ajouter des légendes et des titres
plt.legend(loc="lower left")
# sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.xlabel("Langues")
plt.ylabel("F1-Score")
plt.title("Comparaison d'efficacité entre les modèles XML-Roberta et LangId")

# Faire en sorte que les noms de langage soient lisibles
plt.xticks(rotation=45, ha="right")
# Définir les limites de l'axe Y entre 0.9 et 1
plt.ylim(0.95, 1)
# Afficher le graphique
plt.show()
```
